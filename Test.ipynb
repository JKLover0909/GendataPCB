# VAE for Image Generation
This notebook implements a Variational Autoencoder to learn and generate images similar to those in the "Dị vật mạ" folder.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

# --- CÁC THAM SỐ CẤU HÌNH ---
IMAGE_PATH = '/home/jkl0909/VAE-gendata/raw/Ngắn mạch'
IMG_WIDTH = 128
IMG_HEIGHT = 128
BATCH_SIZE = 16
LATENT_DIM = 64  # Kích thước không gian tiềm ẩn, có thể tinh chỉnh
EPOCHS = 100

# --- 1. TẢI VÀ TIỀN XỬ LÝ DỮ LIỆU ---
def load_and_preprocess_data(path, img_width, img_height):
    image_files = [os.path.join(path, f) for f in os.listdir(path) if f.endswith(('png', 'jpg', 'jpeg'))]
    images = []
    for file in image_files:
        try:
            img = cv2.imread(file)
            if img is not None:
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Chuyển sang RGB
                img = cv2.resize(img, (img_width, img_height))
                images.append(img)
        except Exception as e:
            print(f"Error loading image {file}: {e}")

    if not images:
        raise ValueError("Không tìm thấy ảnh nào trong thư mục hoặc không thể đọc ảnh.")
        
    # Chuẩn hóa ảnh về khoảng [0, 1]
    images = np.array(images, dtype="float32") / 255.0
    print(f"Đã tải thành công {len(images)} ảnh với kích thước: {images.shape}")
    return images

# --- 2. XÂY DỰNG MODEL VAE ---

# Lớp Sampling để thực hiện Reparameterization Trick
class Sampling(layers.Layer):
    """Lấy mẫu từ z_mean và z_log_var (reparameterization trick)."""
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

# Xây dựng ENCODER
encoder_inputs = keras.Input(shape=(IMG_WIDTH, IMG_HEIGHT, 3))
x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Flatten()(x)
x = layers.Dense(128, activation="relu")(x)
z_mean = layers.Dense(LATENT_DIM, name="z_mean")(x)
z_log_var = layers.Dense(LATENT_DIM, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()

# Xây dựng DECODER
latent_inputs = keras.Input(shape=(LATENT_DIM,))
x = layers.Dense(32 * 32 * 64, activation="relu")(latent_inputs)
x = layers.Reshape((32, 32, 64))(x)
x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
decoder_outputs = layers.Conv2DTranspose(3, 3, activation="sigmoid", padding="same")(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()

# Xây dựng VAE model hoàn chỉnh và định nghĩa hàm loss
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            
            # Tính Reconstruction Loss (sai khác giữa ảnh gốc và ảnh tái tạo)
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(data, reconstruction), axis=(1, 2)
                )
            )
            
            # Tính KL Divergence Loss
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))
            
            total_loss = reconstruction_loss + kl_loss
            
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

# --- 3. HUẤN LUYỆN MODEL ---
if __name__ == '__main__':
    # Tải dữ liệu
    pcb_images = load_and_preprocess_data(IMAGE_PATH, IMG_WIDTH, IMG_HEIGHT)

    # Khởi tạo và compile VAE
    vae = VAE(encoder, decoder)
    vae.compile(optimizer=keras.optimizers.Adam())

    # Huấn luyện model
    print("\nBắt đầu quá trình huấn luyện VAE...")
    vae.fit(pcb_images, epochs=EPOCHS, batch_size=BATCH_SIZE)
    print("Huấn luyện hoàn tất!")

    # --- 4. TẠO ẢNH MỚI ---
    print("\nĐang tạo ảnh mới từ không gian tiềm ẩn...")
    
    # Lấy mẫu ngẫu nhiên từ phân phối chuẩn
    random_latent_vector = tf.random.normal(shape=(1, LATENT_DIM))
    
    # Dùng decoder để tạo ảnh
    generated_image_normalized = vae.decoder.predict(random_latent_vector)
    
    # Chuyển đổi ảnh về định dạng có thể hiển thị/lưu
    generated_image = (generated_image_normalized[0] * 255).astype(np.uint8)
    
    # Lưu ảnh
    output_filename = 'generated_pcb_short_circuit.png'
    cv2.imwrite(output_filename, cv2.cvtColor(generated_image, cv2.COLOR_RGB2BGR))
    print(f"Đã lưu ảnh mới tại: {output_filename}")

    # Hiển thị ảnh
    plt.imshow(generated_image)
    plt.title("Generated PCB Short Circuit Image")
    plt.axis("off")
    plt.show()
```

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import glob
from sklearn.model_selection import train_test_split
```

```python
# Load and preprocess images
def load_images_from_folder(folder_path, img_size=(64, 64)):
    """Load images from folder and preprocess them"""
    images = []
    image_paths = glob.glob(os.path.join(folder_path, "*.jpg")) + \
                  glob.glob(os.path.join(folder_path, "*.png")) + \
                  glob.glob(os.path.join(folder_path, "*.jpeg"))
    
    for img_path in image_paths:
        try:
            img = Image.open(img_path).convert('RGB')
            img = img.resize(img_size)
            img_array = np.array(img) / 255.0  # Normalize to [0,1]
            images.append(img_array)
        except Exception as e:
            print(f"Error loading {img_path}: {e}")
    
    return np.array(images)

# Load images
folder_path = "/home/jkl0909/VAE-gendata/raw/Dị vật mạ"
images = load_images_from_folder(folder_path)
print(f"Loaded {len(images)} images with shape {images.shape}")
```

```python
# VAE Model Components
class Sampling(layers.Layer):
    """Sampling layer for VAE"""
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

def build_encoder(input_shape, latent_dim):
    """Build encoder network"""
    encoder_inputs = keras.Input(shape=input_shape)
    x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
    x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
    x = layers.Conv2D(128, 3, activation="relu", strides=2, padding="same")(x)
    x = layers.Flatten()(x)
    x = layers.Dense(16, activation="relu")(x)
    
    z_mean = layers.Dense(latent_dim, name="z_mean")(x)
    z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
    z = Sampling()([z_mean, z_log_var])
    
    encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
    return encoder

def build_decoder(latent_dim, output_shape):
    """Build decoder network"""
    latent_inputs = keras.Input(shape=(latent_dim,))
    x = layers.Dense(8 * 8 * 128, activation="relu")(latent_inputs)
    x = layers.Reshape((8, 8, 128))(x)
    x = layers.Conv2DTranspose(128, 3, activation="relu", strides=2, padding="same")(x)
    x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
    x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
    decoder_outputs = layers.Conv2DTranspose(3, 3, activation="sigmoid", padding="same")(x)
    
    decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
    return decoder
```

```python
# VAE Model
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(data, reconstruction), axis=(1, 2)
                )
            )
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))
            total_loss = reconstruction_loss + kl_loss
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }
```

```python
# Initialize and compile VAE
input_shape = (64, 64, 3)
latent_dim = 16

encoder = build_encoder(input_shape, latent_dim)
decoder = build_decoder(latent_dim, input_shape)

# Create VAE
vae = VAE(encoder, decoder)
vae.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001))

# Display model architectures
print("Encoder architecture:")
encoder.summary()
print("\nDecoder architecture:")
decoder.summary()
```

```python
# Train the VAE
if len(images) > 0:
    # Split data if we have enough images
    if len(images) > 10:
        train_images, val_images = train_test_split(images, test_size=0.2, random_state=42)
    else:
        train_images = images
        val_images = images
    
    # Train the model
    history = vae.fit(
        train_images,
        epochs=50,
        batch_size=8,
        validation_data=(val_images, None) if len(val_images) > 0 else None,
        verbose=1
    )
    
    # Plot training history
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1)
    plt.plot(history.history['loss'])
    if 'val_loss' in history.history:
        plt.plot(history.history['val_loss'])
    plt.title('Total Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'] if 'val_loss' in history.history else ['Train'])
    
    plt.subplot(1, 3, 2)
    plt.plot(history.history['reconstruction_loss'])
    if 'val_reconstruction_loss' in history.history:
        plt.plot(history.history['val_reconstruction_loss'])
    plt.title('Reconstruction Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'] if 'val_reconstruction_loss' in history.history else ['Train'])
    
    plt.subplot(1, 3, 3)
    plt.plot(history.history['kl_loss'])
    if 'val_kl_loss' in history.history:
        plt.plot(history.history['val_kl_loss'])
    plt.title('KL Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'] if 'val_kl_loss' in history.history else ['Train'])
    
    plt.tight_layout()
    plt.show()
else:
    print("No images found in the specified folder!")
```

```python
# Generate new images
def generate_images(vae, num_images=6):
    """Generate new images using the trained VAE"""
    # Sample random points in the latent space
    random_latent_vectors = tf.random.normal(shape=(num_images, latent_dim))
    
    # Generate images
    generated_images = vae.decoder(random_latent_vectors)
    
    # Display generated images
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))
    axes = axes.flatten()
    
    for i in range(num_images):
        axes[i].imshow(generated_images[i])
        axes[i].axis('off')
        axes[i].set_title(f'Generated Image {i+1}')
    
    plt.tight_layout()
    plt.show()
    
    return generated_images

# Generate and display new images
if len(images) > 0:
    print("Generating new images...")
    generated_images = generate_images(vae, num_images=6)
else:
    print("Cannot generate images - no training data available!")
```

```python
# Display original vs reconstructed images
def show_reconstruction_comparison(vae, original_images, num_samples=3):
    """Compare original images with their reconstructions"""
    if len(original_images) == 0:
        print("No images to compare!")
        return
    
    # Select random samples
    indices = np.random.choice(len(original_images), min(num_samples, len(original_images)), replace=False)
    sample_images = original_images[indices]
    
    # Get reconstructions
    _, _, z = vae.encoder(sample_images)
    reconstructions = vae.decoder(z)
    
    # Display comparison
    fig, axes = plt.subplots(2, num_samples, figsize=(15, 6))
    
    for i in range(num_samples):
        # Original images
        axes[0, i].imshow(sample_images[i])
        axes[0, i].set_title(f'Original {i+1}')
        axes[0, i].axis('off')
        
        # Reconstructed images
        axes[1, i].imshow(reconstructions[i])
        axes[1, i].set_title(f'Reconstructed {i+1}')
        axes[1, i].axis('off')
    
    plt.tight_layout()
    plt.show()

# Show reconstruction comparison
if len(images) > 0:
    show_reconstruction_comparison(vae, images, num_samples=min(3, len(images)))
```

```python
# Save generated images
def save_generated_images(generated_images, save_dir="/home/jkl0909/VAE-gendata/generated"):
    """Save generated images to disk"""
    os.makedirs(save_dir, exist_ok=True)
    
    for i, img in enumerate(generated_images):
        # Convert to PIL Image
        img_array = (img.numpy() * 255).astype(np.uint8)
        pil_img = Image.fromarray(img_array)
        
        # Save image
        save_path = os.path.join(save_dir, f"generated_image_{i+1}.png")
        pil_img.save(save_path)
        print(f"Saved: {save_path}")

# Save the generated images
if len(images) > 0:
    generated_images = generate_images(vae, num_images=10)
    save_generated_images(generated_images)
    print("Generated images saved to /home/jkl0909/VAE-gendata/generated/")
```
